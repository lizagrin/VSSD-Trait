## Локальная копия датасета и manifests

```
import subprocess

def rsync_dir(src: str, dst: str, info: bool = True):
    os.makedirs(dst, exist_ok=True)
    cmd = ["rsync", "-ah", "--partial", "--append-verify"]
    if info:
        cmd.append("--info=progress2")
    cmd.extend([src.rstrip("/") + "/", dst.rstrip("/") + "/"])
    print("RUN:", " ".join(cmd))
    subprocess.run(cmd, check=True)

def copy_dataset_from_drive(overwrite: bool = False):
    if overwrite and os.path.exists(cfg.local_dataset_root):
        shutil.rmtree(cfg.local_dataset_root)
        os.makedirs(cfg.local_dataset_root, exist_ok=True)
    rsync_dir(cfg.drive_dataset_root, cfg.local_dataset_root, info=True)
    print("Local dataset ready at:", cfg.local_dataset_root)

def load_pickle_compat(path):
    with open(path, "rb") as f:
        try:
            return pickle.load(f)
        except UnicodeDecodeError:
            f.seek(0)
            return pickle.load(f, encoding="latin1")

def get_split_paths(root, split):
    split_map = {
        "train": {"dir": "TRAIN", "ann": "annotation_training.pkl", "tr": "transcription_training.pkl"},
        "validation": {"dir": "VALIDATION", "ann": "annotation_validation.pkl", "tr": "transcription_validation.pkl"},
        "test": {"dir": "TEST", "ann": "annotation_test.pkl", "tr": "transcription_test.pkl"},
    }
    spec = split_map[split]
    split_dir = os.path.join(root, spec["dir"])
    ann_path = os.path.join(split_dir, "Annotation", spec["ann"])
    tr_path = os.path.join(split_dir, "Annotation", spec["tr"])
    if not os.path.exists(ann_path):
        raise FileNotFoundError(f"Не найден файл аннотаций: {ann_path}")
    if not os.path.exists(tr_path):
        tr_path = None
    return split_dir, ann_path, tr_path

def possible_video_keys(video_name):
    stem = Path(video_name).stem
    fname = Path(video_name).name
    return [stem, fname]

def get_by_video_name(mapping, video_name):
    if not isinstance(mapping, dict):
        return None
    for key in possible_video_keys(video_name):
        if key in mapping:
            return mapping[key]
    return None

def get_trait_value(annotation_obj, trait_name, video_name):
    if not isinstance(annotation_obj, dict) or trait_name not in annotation_obj:
        return None
    return get_by_video_name(annotation_obj[trait_name], video_name)

def get_transcription_value(transcription_obj, video_name):
    if transcription_obj is None:
        return ""
    if isinstance(transcription_obj, dict):
        val = get_by_video_name(transcription_obj, video_name)
        if val is not None:
            return val if isinstance(val, str) else str(val)
        if "transcription" in transcription_obj and isinstance(transcription_obj["transcription"], dict):
            val = get_by_video_name(transcription_obj["transcription"], video_name)
            if val is not None:
                return val if isinstance(val, str) else str(val)
    return ""

def build_manifest_from_local(root):
    manifests = {}
    for split in ["train", "validation", "test"]:
        split_dir, ann_path, tr_path = get_split_paths(root, split)
        ann = load_pickle_compat(ann_path)
        tr = load_pickle_compat(tr_path) if tr_path is not None else None

        video_files = []
        for ext in ["*.mp4", "*.avi", "*.mov", "*.mkv", "*.webm"]:
            video_files.extend(glob.glob(os.path.join(split_dir, ext)))

        rows = []
        skipped = 0
        for video_path in sorted(video_files):
            video_name = os.path.basename(video_path)
            row = {
                "split": split,
                "video_id": Path(video_path).stem,
                "video_path": video_path,
                "transcription": get_transcription_value(tr, video_name),
            }
            ok = True
            for trait in cfg.trait_names:
                value = get_trait_value(ann, trait, video_name)
                if value is None:
                    ok = False
                    break
                row[trait] = float(value)
            if ok:
                rows.append(row)
            else:
                skipped += 1

        manifests[split] = pd.DataFrame(rows)
        print(f"{split}: {manifests[split].shape}, skipped={skipped}")
    return manifests

# 1) copy DATA from Drive to local Colab VM
copy_dataset_from_drive(overwrite=False)

# 2) build manifests from local copy
manifests = build_manifest_from_local(cfg.local_dataset_root)

train_manifest = manifests["train"].copy().reset_index(drop=True)
val_manifest = manifests["validation"].copy().reset_index(drop=True)
test_manifest = manifests["test"].copy().reset_index(drop=True)

if cfg.max_train_samples is not None:
    train_manifest = train_manifest.sample(min(cfg.max_train_samples, len(train_manifest)), random_state=SEED).reset_index(drop=True)
if cfg.max_val_samples is not None:
    val_manifest = val_manifest.sample(min(cfg.max_val_samples, len(val_manifest)), random_state=SEED).reset_index(drop=True)
if cfg.max_test_samples is not None:
    test_manifest = test_manifest.sample(min(cfg.max_test_samples, len(test_manifest)), random_state=SEED).reset_index(drop=True)

display(train_manifest.head())
print(train_manifest.shape, val_manifest.shape, test_manifest.shape)
```

## Предобработка, быстрый локальный кэш и промежуточный sync в Drive

```
VIDEO_CACHE_DIR = os.path.join(cfg.cache_root, "video")
AUDIO_CACHE_DIR = os.path.join(cfg.cache_root, "audio")
TEXT_CACHE_DIR = os.path.join(cfg.cache_root, "text")
for p in [VIDEO_CACHE_DIR, AUDIO_CACHE_DIR, TEXT_CACHE_DIR]:
    os.makedirs(p, exist_ok=True)

mtcnn = MTCNN(
    image_size=cfg.face_size,
    margin=18,
    keep_all=True,
    post_process=False,
    device=DEVICE,
)

_text_model = None

def get_text_encoder():
    global _text_model
    if _text_model is None and cfg.use_text_branch:
        _text_model = SentenceTransformer(cfg.text_model_name, device=DEVICE if DEVICE == "cuda" else "cpu")
    return _text_model

def sanitize_text(text):
    if text is None:
        return ""
    text = str(text).strip()
    text = re.sub(r"\s+", " ", text)
    return text

def cache_subdir(mode: str):
    stride = {"train": cfg.train_stride_sec, "eval": cfg.eval_stride_sec, "dense": cfg.dense_stride_sec, "test": cfg.eval_stride_sec}[mode]
    return f"{cfg.cache_version}_f{cfg.face_size}_k{cfg.frames_per_window}_mel{cfg.mel_frames_per_window}_s{stride}_txt{int(cfg.use_text_branch)}"

def get_cache_path(split: str, video_id: str, mode: str):
    folder = os.path.join(VIDEO_CACHE_DIR, cache_subdir(mode), split)
    os.makedirs(folder, exist_ok=True)
    return os.path.join(folder, f"{video_id}.pt")

def safe_video_reader(video_path):
    try:
        return VideoReader(video_path, ctx=cpu(0))
    except Exception as e:
        print(f"Video open error: {video_path} -> {e}")
        return None

def build_window_starts(duration_sec: float, mode: str):
    stride = {
        "train": cfg.train_stride_sec,
        "eval": cfg.eval_stride_sec,
        "dense": cfg.dense_stride_sec,
        "test": cfg.eval_stride_sec,
    }[mode]
    if duration_sec <= cfg.window_sec + 1e-6:
        return [max(0.0, (duration_sec - cfg.window_sec) / 2.0)]
    starts = np.arange(0.0, max(0.0, duration_sec - cfg.window_sec) + 1e-6, stride).tolist()
    if starts[-1] < duration_sec - cfg.window_sec - 1e-3:
        starts.append(max(0.0, duration_sec - cfg.window_sec))
    return [float(x) for x in starts]

def sample_times_in_window(start_sec, end_sec, n_frames):
    eps = (end_sec - start_sec) / (2 * max(n_frames, 1))
    left = start_sec + eps
    right = max(left, end_sec - eps)
    return np.linspace(left, right, n_frames).tolist()

def seconds_to_idx(sec, fps, n_frames_total):
    idx = int(round(sec * fps))
    return max(0, min(n_frames_total - 1, idx))

def _largest_box(boxes, probs):
    if boxes is None or len(boxes) == 0:
        return None
    if probs is None:
        probs = np.ones(len(boxes))
    areas = (boxes[:, 2] - boxes[:, 0]) * (boxes[:, 3] - boxes[:, 1])
    score = probs + 1e-3 * (areas / np.maximum(areas.max(), 1.0))
    return boxes[int(np.argmax(score))]

def crop_from_box(frame_rgb, box):
    h, w = frame_rgb.shape[:2]
    x1, y1, x2, y2 = box
    x1, y1, x2, y2 = map(float, [x1, y1, x2, y2])
    bw = max(1.0, x2 - x1)
    bh = max(1.0, y2 - y1)
    cx = (x1 + x2) / 2.0
    cy = (y1 + y2) / 2.0
    side = max(bw, bh) * 1.35
    x1 = int(round(max(0, cx - side / 2)))
    y1 = int(round(max(0, cy - side / 2)))
    x2 = int(round(min(w, cx + side / 2)))
    y2 = int(round(min(h, cy + side / 2)))
    crop = frame_rgb[y1:y2, x1:x2]
    if crop.size == 0:
        side = min(h, w)
        y1 = (h - side) // 2
        x1 = (w - side) // 2
        crop = frame_rgb[y1:y1+side, x1:x1+side]
    crop = cv2.resize(crop, (cfg.face_size, cfg.face_size))
    return crop

def detect_face_or_track(frame_rgb, prev_box=None):
    try:
        boxes, probs = mtcnn.detect(frame_rgb)
        box = _largest_box(boxes, probs)
    except Exception:
        box = None

    if box is None and prev_box is not None:
        crop = crop_from_box(frame_rgb, prev_box)
        return crop, prev_box, 0.5

    if box is not None and prev_box is not None:
        box = 0.75 * np.array(box) + 0.25 * np.array(prev_box)

    if box is not None:
        crop = crop_from_box(frame_rgb, box)
        return crop, box, 1.0

    h, w = frame_rgb.shape[:2]
    side = min(h, w)
    y1 = (h - side) // 2
    x1 = (w - side) // 2
    crop = frame_rgb[y1:y1+side, x1:x1+side]
    crop = cv2.resize(crop, (cfg.face_size, cfg.face_size))
    return crop, None, 0.0

def load_audio_from_video(video_path, sample_rate=16000):
    wav_path = os.path.join(AUDIO_CACHE_DIR, Path(video_path).stem + ".wav")
    if not os.path.exists(wav_path):
        cmd = f'ffmpeg -y -i "{video_path}" -ac 1 -ar {sample_rate} "{wav_path}" -loglevel error'
        os.system(cmd)
    y, sr = librosa.load(wav_path, sr=sample_rate, mono=True)
    return y, sr, wav_path

def make_logmel(y, sr):
    mel = librosa.feature.melspectrogram(
        y=y, sr=sr, n_mels=cfg.n_mels, n_fft=cfg.n_fft,
        hop_length=cfg.hop_length, win_length=cfg.win_length, power=2.0,
    )
    mel = librosa.power_to_db(mel, ref=np.max)
    mel = (mel - mel.mean()) / (mel.std() + 1e-6)
    return mel.astype(np.float32)

def make_window_mels(y, sr, starts):
    out = []
    seg_len = int(round(cfg.window_sec * sr))
    for s in starts:
        s_idx = int(round(s * sr))
        e_idx = s_idx + seg_len
        seg = y[s_idx:e_idx]
        if len(seg) < seg_len:
            seg = np.pad(seg, (0, seg_len - len(seg)))
        mel = make_logmel(seg, sr)
        if mel.shape[1] < cfg.mel_frames_per_window:
            mel = np.pad(mel, ((0, 0), (0, cfg.mel_frames_per_window - mel.shape[1])))
        mel = mel[:, :cfg.mel_frames_per_window]
        out.append(mel)
    return np.stack(out).astype(np.float32)

def make_basic_prosody(y, sr):
    rms = librosa.feature.rms(y=y)[0]
    zcr = librosa.feature.zero_crossing_rate(y)[0]
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
    try:
        f0 = librosa.yin(y, fmin=50, fmax=400, sr=sr)
    except Exception:
        f0 = np.zeros_like(rms)
    feats = []
    for arr in [rms, zcr, f0]:
        arr = np.nan_to_num(arr, nan=0.0, posinf=0.0, neginf=0.0)
        feats.extend([arr.mean(), arr.std(), np.median(arr), arr.min(), arr.max()])
    mfcc = np.nan_to_num(mfcc, nan=0.0, posinf=0.0, neginf=0.0)
    feats.extend(mfcc.mean(axis=1).tolist())
    feats.extend(mfcc.std(axis=1).tolist())
    return np.array(feats, dtype=np.float32)

smile = opensmile.Smile(
    feature_set=opensmile.FeatureSet.eGeMAPSv02,
    feature_level=opensmile.FeatureLevel.Functionals,
) if cfg.use_ege_maps else None

def make_ege_maps(wav_path):
    if smile is None:
        return None
    try:
        df = smile.process_file(wav_path)
        return df.iloc[0].values.astype(np.float32)
    except Exception:
        return None

def encode_text_cached(video_id: str, text: str):
    text = sanitize_text(text)
    cache_path = os.path.join(
        TEXT_CACHE_DIR,
        f"{video_id}_{cfg.text_model_name.split('/')[-1]}.npy"
    )

    if os.path.exists(cache_path):
        return np.load(cache_path).astype(np.float32)

    if not cfg.use_text_branch:
        return np.zeros(cfg.text_emb_dim, dtype=np.float32)

    if len(text) == 0:
        emb = np.zeros(cfg.text_emb_dim, dtype=np.float32)
    else:
        model = get_text_encoder()
        emb = model.encode([text], normalize_embeddings=True, convert_to_numpy=True)[0].astype(np.float32)

    tmp_path = cache_path + ".tmp"
    with open(tmp_path, "wb") as f:
        np.save(f, emb)
    os.replace(tmp_path, cache_path)

    return emb

def sync_local_cache_to_drive_safe(retries: int = 3, sleep_sec: int = 10):
    os.makedirs(cfg.drive_cache_root, exist_ok=True)

    cmd = [
        "rsync", "-ah",
        "--partial",
        "--append-verify",
        cfg.cache_root.rstrip("/") + "/",
        cfg.drive_cache_root.rstrip("/") + "/",
    ]

    for attempt in range(1, retries + 1):
        print(f"RUN sync ({attempt}/{retries}):", " ".join(cmd))
        result = subprocess.run(cmd, capture_output=True, text=True)

        if result.returncode == 0:
            print("sync ok")
            return True

        print("sync failed with code:", result.returncode)
        if result.stderr:
            print(result.stderr[-2000:])

        if attempt < retries:
            time.sleep(sleep_sec)

    print("WARNING: sync failed, but precache will continue")
    return False

def restore_cache_from_drive():
    if not os.path.exists(cfg.drive_cache_root):
        print("Drive cache root does not exist yet:", cfg.drive_cache_root)
        return
    rsync_dir(cfg.drive_cache_root, cfg.cache_root, info=True)
    print("Local cache restored from Drive.")

def save_cache_progress(name: str, done: int, total: int):
    os.makedirs(cfg.logs_dir, exist_ok=True)
    progress_path = os.path.join(cfg.logs_dir, f"cache_progress_{name}.json")
    with open(progress_path, "w", encoding="utf-8") as f:
        json.dump({"done": int(done), "total": int(total)}, f, ensure_ascii=False, indent=2)

def preprocess_and_cache_video(row, mode="train", overwrite=False):
    split = row["split"]
    video_id = row["video_id"]
    video_path = row["video_path"]
    cache_path = get_cache_path(split, video_id, mode)

    if os.path.exists(cache_path) and not overwrite:
        return cache_path

    vr = safe_video_reader(video_path)
    if vr is None or len(vr) == 0:
        return None

    n_frames_total = len(vr)
    fps = float(vr.get_avg_fps()) if hasattr(vr, "get_avg_fps") else 25.0
    duration_sec = n_frames_total / max(fps, 1e-6)
    starts = build_window_starts(duration_sec, mode)

    all_faces = []
    all_scores = []
    prev_box = None

    for s in starts:
        e = min(duration_sec, s + cfg.window_sec)
        times = sample_times_in_window(s, e, cfg.frames_per_window)
        faces = []
        scores = []
        for t in times:
            idx = seconds_to_idx(t, fps, n_frames_total)
            frame = vr[idx].asnumpy()
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            face, prev_box, score = detect_face_or_track(frame_rgb, prev_box=prev_box)
            faces.append(face)
            scores.append(score)
        all_faces.append(np.stack(faces))
        all_scores.append(np.array(scores, dtype=np.float32))

    y, sr, wav_path = load_audio_from_video(video_path, sample_rate=cfg.sample_rate)
    mel_windows = make_window_mels(y, sr, starts)
    prosody = make_basic_prosody(y, sr)
    ege = make_ege_maps(wav_path)
    text_emb = encode_text_cached(video_id, row.get("transcription", ""))

    target = np.array([row[t] for t in cfg.trait_names], dtype=np.float32)

    payload = {
        "video_id": video_id,
        "video_path": video_path,
        "split": split,
        "faces": np.stack(all_faces).astype(np.uint8),
        "face_scores": np.stack(all_scores).astype(np.float32),
        "mel_windows": mel_windows.astype(np.float32),
        "prosody": prosody.astype(np.float32),
        "ege_maps": ege.astype(np.float32) if ege is not None else None,
        "text_emb": text_emb.astype(np.float32),
        "target": target,
        "transcription": row.get("transcription", ""),
    }

    tmp_path = cache_path + ".tmp"
    torch.save(payload, tmp_path)
    os.replace(tmp_path, cache_path)
    return cache_path

def precache_manifest(manifest, mode="train", overwrite=False, sync_every=None, name="split"):
    ok = 0

    for i, (_, row) in enumerate(tqdm(manifest.iterrows(), total=len(manifest), desc=f"cache:{name}"), start=1):
        try:
            path = preprocess_and_cache_video(row, mode=mode, overwrite=overwrite)
            if path is not None:
                ok += 1
        except Exception as e:
            print(f"\nERROR on video_id={row.get('video_id')} : {e}")
            continue

        if sync_every is not None and i % sync_every == 0:
            sync_local_cache_to_drive_safe(retries=3, sleep_sec=15)

    # финальный sync
    sync_local_cache_to_drive_safe(retries=5, sleep_sec=20)
    print(f"cached {ok}/{len(manifest)} for mode={mode}")

print("train cache folder:", os.path.join(VIDEO_CACHE_DIR, cache_subdir("train"), "train"))
print("eval  cache folder:", os.path.join(VIDEO_CACHE_DIR, cache_subdir("eval"), "validation"))
print("test  cache folder:", os.path.join(VIDEO_CACHE_DIR, cache_subdir("test"), "test"))
print("dense cache folder:", os.path.join(VIDEO_CACHE_DIR, cache_subdir("dense"), "validation"))
```

```
class FaceAugment:
    def __init__(self, train=True):
        self.train = train

    def __call__(self, x):
        img = x.astype(np.uint8)
        if self.train:
            if random.random() < 0.5:
                img = cv2.flip(img, 1)
            if random.random() < 0.4:
                alpha = 1.0 + random.uniform(-0.10, 0.10)
                beta = random.uniform(-10, 10)
                img = np.clip(img.astype(np.float32) * alpha + beta, 0, 255).astype(np.uint8)
            if random.random() < 0.25:
                img = cv2.GaussianBlur(img, (3, 3), 0)
        img = img.astype(np.float32) / 255.0
        img = img.transpose(2, 0, 1)
        return torch.tensor(img, dtype=torch.float32)


class FIv2CacheDataset(Dataset):
    def __init__(self, manifest: pd.DataFrame, mode="train", cache_only=True):
        self.manifest = manifest.reset_index(drop=True)
        self.mode = mode
        self.train = mode == "train"
        self.cache_only = cache_only
        self.face_tf = FaceAugment(train=self.train)

    def __len__(self):
        return len(self.manifest)

    def __getitem__(self, idx):
        row = self.manifest.iloc[idx]
        cache_path = get_cache_path(row["split"], row["video_id"], self.mode)

        # НИЧЕГО не догружаем из видео, работаем только с готовым кэшем
        if not os.path.exists(cache_path):
            print(f"[SKIP missing cache] {row['video_id']} -> {cache_path}")
            return None

        try:
            obj = torch.load(cache_path, map_location="cpu", weights_only=False)
        except Exception as e:
            print(f"[SKIP broken cache] {row['video_id']} -> {cache_path} | {repr(e)}")
            return None

        try:
            faces = obj["faces"]
            face_scores = obj["face_scores"]
            mel_windows = obj["mel_windows"]
            prosody = obj["prosody"]
            ege = obj["ege_maps"]
            text_emb = obj["text_emb"]
            target = obj["target"]

            faces = torch.stack([
                torch.stack([self.face_tf(faces[w, k]) for k in range(faces.shape[1])], dim=0)
                for w in range(faces.shape[0])
            ], dim=0)

            out = {
                "video_id": row["video_id"],
                "faces": faces,
                "face_scores": torch.tensor(face_scores, dtype=torch.float32),
                "mel_windows": torch.tensor(mel_windows, dtype=torch.float32),
                "prosody": torch.tensor(prosody, dtype=torch.float32),
                "ege_maps": torch.tensor(
                    np.zeros(88, dtype=np.float32) if ege is None else ege,
                    dtype=torch.float32
                ),
                "text_emb": torch.tensor(text_emb, dtype=torch.float32),
                "target": torch.tensor(target, dtype=torch.float32),
                "num_windows": int(faces.shape[0]),
            }
            return out

        except Exception as e:
            print(f"[SKIP malformed cache] {row['video_id']} -> {cache_path} | {repr(e)}")
            return None
def collate_fn(batch):
    # выбрасываем битые/отсутствующие элементы
    batch = [x for x in batch if x is not None]

    # если весь батч оказался битым — возвращаем None
    if len(batch) == 0:
        return None

    B = len(batch)
    Wmax = max(x["num_windows"] for x in batch)
    K = batch[0]["faces"].shape[1]
    C = batch[0]["faces"].shape[2]
    H = batch[0]["faces"].shape[3]
    Wimg = batch[0]["faces"].shape[4]
    n_mels = batch[0]["mel_windows"].shape[1]
    Tm = batch[0]["mel_windows"].shape[2]

    faces = torch.zeros(B, Wmax, K, C, H, Wimg, dtype=torch.float32)
    face_scores = torch.zeros(B, Wmax, K, dtype=torch.float32)
    mel_windows = torch.zeros(B, Wmax, n_mels, Tm, dtype=torch.float32)
    window_mask = torch.zeros(B, Wmax, dtype=torch.bool)

    prosody = []
    ege_maps = []
    text_emb = []
    target = []
    video_ids = []

    for i, x in enumerate(batch):
        w = x["num_windows"]
        faces[i, :w] = x["faces"]
        face_scores[i, :w] = x["face_scores"]
        mel_windows[i, :w] = x["mel_windows"]
        window_mask[i, :w] = True

        prosody.append(x["prosody"])
        ege_maps.append(x["ege_maps"])
        text_emb.append(x["text_emb"])
        target.append(x["target"])
        video_ids.append(x["video_id"])

    return {
        "video_id": video_ids,
        "faces": faces,
        "face_scores": face_scores,
        "mel_windows": mel_windows,
        "window_mask": window_mask,
        "prosody": torch.stack(prosody),
        "ege_maps": torch.stack(ege_maps),
        "text_emb": torch.stack(text_emb),
        "target": torch.stack(target),
    }

def make_loader(manifest, mode="train", shuffle=None, cache_only=True, num_workers=None):
    ds = FIv2CacheDataset(manifest, mode=mode, cache_only=cache_only)

    loader = DataLoader(
        ds,
        batch_size=cfg.batch_size if mode == "train" else cfg.eval_batch_size,
        shuffle=(mode == "train") if shuffle is None else shuffle,
        num_workers=0 if num_workers is None else num_workers,
        pin_memory=True,
        collate_fn=collate_fn,
        drop_last=False,
    )
    return ds, loader

train_ds, train_loader = make_loader(
    train_manifest_cached,
    mode="train",
    cache_only=True,
    num_workers=0,
)

val_ds, val_loader = make_loader(
    val_manifest_cached,
    mode="eval",
    cache_only=True,
    num_workers=0,
)

# test может быть пустым — это нормально
if len(test_manifest_cached) > 0:
    test_ds, test_loader = make_loader(
        test_manifest_cached,
        mode="test",
        cache_only=True,
        num_workers=0,
    )
    print("loaders:", len(train_loader), len(val_loader), len(test_loader))
else:
    test_ds, test_loader = None, None
    print("loaders:", len(train_loader), len(val_loader), "test skipped (no cache)")
```


## Модель: window-level audiovisual fusion + optional text branch


```
class AttentionPool(nn.Module):
    def __init__(self, dim):
        super().__init__()
        self.score = nn.Linear(dim, 1)

    def forward(self, x, mask=None):
        logits = self.score(x).squeeze(-1)

        if mask is not None:
            mask = mask.bool()

            # считаем softmax в float32, чтобы не ловить overflow в half
            logits_fp32 = logits.float()
            logits_fp32 = logits_fp32.masked_fill(
                ~mask,
                torch.finfo(logits_fp32.dtype).min
            )
            attn = torch.softmax(logits_fp32, dim=-1).to(x.dtype)
        else:
            attn = torch.softmax(logits.float(), dim=-1).to(x.dtype)

        pooled = torch.einsum("bn,bnd->bd", attn, x)
        return pooled, attn

class AudioDeepCNN(nn.Module):
    def __init__(self, out_dim=256, p=0.1):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv2d(1, 32, 3, 1, 1),
            nn.BatchNorm2d(32),
            nn.GELU(),
            nn.MaxPool2d(2),
            nn.Conv2d(32, 64, 3, 1, 1),
            nn.BatchNorm2d(64),
            nn.GELU(),
            nn.MaxPool2d(2),
            nn.Conv2d(64, 128, 3, 1, 1),
            nn.BatchNorm2d(128),
            nn.GELU(),
            nn.AdaptiveAvgPool2d((1, 1)),
            nn.Flatten(),
            nn.Dropout(p),
            nn.Linear(128, out_dim),
        )

    def forward(self, mel):
        return self.net(mel.unsqueeze(1))

class HCBranch(nn.Module):
    def __init__(self, in_dim, hidden_dim=128, out_dim=128, p=0.1):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(in_dim, hidden_dim),
            nn.LayerNorm(hidden_dim),
            nn.GELU(),
            nn.Dropout(p),
            nn.Linear(hidden_dim, out_dim),
        )

    def forward(self, x):
        return self.net(x)

class VisualTemporalModule(nn.Module):
    def __init__(self, backbone, visual_dim, hidden_dim=512, p=0.15):
        super().__init__()
        self.backbone = backbone
        self.frame_norm = nn.LayerNorm(visual_dim)
        self.frame_pool = AttentionPool(visual_dim)
        self.window_proj = nn.Sequential(
            nn.Linear(visual_dim + 1, hidden_dim),
            nn.LayerNorm(hidden_dim),
            nn.GELU(),
            nn.Dropout(p),
        )
        self.bigru = nn.GRU(
            input_size=hidden_dim,
            hidden_size=hidden_dim // 2,
            num_layers=2,
            batch_first=True,
            bidirectional=True,
            dropout=p,
        )
        self.self_attn = nn.MultiheadAttention(hidden_dim, num_heads=8, dropout=p, batch_first=True)
        self.temporal_ln = nn.LayerNorm(hidden_dim)
        self.window_pool = AttentionPool(hidden_dim)
        self.out = nn.Sequential(
            nn.Linear(hidden_dim, hidden_dim),
            nn.LayerNorm(hidden_dim),
            nn.GELU(),
            nn.Dropout(p),
        )

    def forward(self, faces, face_scores, window_mask):
        B, W, K, C, H, Wimg = faces.shape
        x = faces.view(B * W * K, C, H, Wimg)
        feats = self.backbone.forward_features(x)
        D = feats.shape[-1]
        feats = feats.view(B, W, K, D)
        feats = self.frame_norm(feats)

        frame_tokens = feats.view(B * W, K, D)
        pooled_frames, frame_attn = self.frame_pool(frame_tokens)
        pooled_frames = pooled_frames.view(B, W, D)
        frame_attn = frame_attn.view(B, W, K)

        fs = face_scores.mean(dim=-1, keepdim=True)
        win_tokens = torch.cat([pooled_frames, fs], dim=-1)
        win_tokens = self.window_proj(win_tokens)

        lengths = window_mask.sum(dim=1).cpu()
        packed = pack_padded_sequence(win_tokens, lengths=lengths, batch_first=True, enforce_sorted=False)
        packed_out, _ = self.bigru(packed)
        temp, _ = pad_packed_sequence(packed_out, batch_first=True, total_length=W)

        attn_out, _ = self.self_attn(temp, temp, temp, key_padding_mask=~window_mask, need_weights=False)
        temp = self.temporal_ln(temp + attn_out)

        clip_vec, window_attn = self.window_pool(temp, mask=window_mask)
        clip_vec = self.out(clip_vec)

        return clip_vec, temp, {
            "frame_attn": frame_attn,
            "window_attn": window_attn,
            "frame_feats": feats,
        }

class AudioWindowModule(nn.Module):
    def __init__(self, prosody_dim, ege_dim, hidden_dim=512, p=0.15):
        super().__init__()
        self.deep = AudioDeepCNN(out_dim=cfg.audio_deep_dim, p=p)
        self.hc = HCBranch(prosody_dim + ege_dim, hidden_dim=cfg.audio_hc_dim, out_dim=cfg.audio_hc_dim, p=p)
        self.window_fuse = nn.Sequential(
            nn.Linear(cfg.audio_deep_dim + cfg.audio_hc_dim, hidden_dim),
            nn.LayerNorm(hidden_dim),
            nn.GELU(),
            nn.Dropout(p),
        )
        self.temporal = nn.MultiheadAttention(hidden_dim, num_heads=8, dropout=p, batch_first=True)
        self.temporal_ln = nn.LayerNorm(hidden_dim)
        self.window_pool = AttentionPool(hidden_dim)

    def forward(self, mel_windows, prosody, ege_maps, window_mask):
        B, W, M, T = mel_windows.shape
        deep = self.deep(mel_windows.view(B * W, M, T)).view(B, W, -1)
        hc = self.hc(torch.cat([prosody, ege_maps], dim=-1))
        hc_expand = hc.unsqueeze(1).expand(-1, W, -1)
        a_tokens = self.window_fuse(torch.cat([deep, hc_expand], dim=-1))
        attn_out, _ = self.temporal(a_tokens, a_tokens, a_tokens, key_padding_mask=~window_mask, need_weights=False)
        a_tokens = self.temporal_ln(a_tokens + attn_out)
        a_clip, a_attn = self.window_pool(a_tokens, mask=window_mask)
        return a_clip, a_tokens, {"audio_deep": deep, "audio_hc": hc, "audio_window_attn": a_attn}

class TextModule(nn.Module):
    def __init__(self, in_dim, out_dim, p=0.10):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(in_dim, out_dim),
            nn.LayerNorm(out_dim),
            nn.GELU(),
            nn.Dropout(p),
        )

    def forward(self, x):
        return self.net(x)

class WindowFusionBlock(nn.Module):
    def __init__(self, dim, p=0.15):
        super().__init__()
        self.fuse = nn.Sequential(
            nn.Linear(dim * 4, dim),
            nn.LayerNorm(dim),
            nn.GELU(),
            nn.Dropout(p),
        )
        self.self_attn = nn.MultiheadAttention(dim, num_heads=8, dropout=p, batch_first=True)
        self.ln = nn.LayerNorm(dim)
        self.pool = AttentionPool(dim)

    def forward(self, v_tokens, a_tokens, window_mask):
        x = torch.cat([v_tokens, a_tokens, torch.abs(v_tokens - a_tokens), v_tokens * a_tokens], dim=-1)
        x = self.fuse(x)
        attn_out, _ = self.self_attn(x, x, x, key_padding_mask=~window_mask, need_weights=False)
        x = self.ln(x + attn_out)
        clip, attn = self.pool(x, mask=window_mask)
        return clip, x, attn

class TraitWiseFusion(nn.Module):
    def __init__(self, modal_dims: List[int], num_traits: int):
        super().__init__()
        self.modal_heads = nn.ModuleList([nn.Linear(d, num_traits) for d in modal_dims])
        self.gate = nn.Linear(sum(modal_dims), num_traits * len(modal_dims))
        self.num_traits = num_traits
        self.num_modalities = len(modal_dims)

    def forward(self, modal_vecs: List[torch.Tensor]):
        preds = [torch.sigmoid(head(vec)) for head, vec in zip(self.modal_heads, modal_vecs)]
        joint = torch.cat(modal_vecs, dim=-1)
        gates = self.gate(joint).view(joint.shape[0], self.num_traits, self.num_modalities)
        alpha = torch.softmax(gates, dim=-1)
        fused = 0.0
        for i, p in enumerate(preds):
            fused = fused + alpha[..., i] * p
        return fused, preds, alpha

class BinHead(nn.Module):
    def __init__(self, in_dim, num_traits=5, num_bins=3):
        super().__init__()
        self.fc = nn.Linear(in_dim, num_traits * num_bins)
        self.num_traits = num_traits
        self.num_bins = num_bins

    def forward(self, x):
        return self.fc(x).view(-1, self.num_traits, self.num_bins)

class AVTPersonalityModel(nn.Module):
    def __init__(self, backbone, visual_dim, prosody_dim, ege_dim):
        super().__init__()
        self.visual = VisualTemporalModule(backbone, visual_dim, hidden_dim=cfg.hidden_dim, p=cfg.dropout)
        self.audio = AudioWindowModule(prosody_dim, ege_dim, hidden_dim=cfg.hidden_dim, p=cfg.dropout)
        self.av_fusion = WindowFusionBlock(cfg.hidden_dim, p=cfg.dropout)

        self.use_text = cfg.use_text_branch
        if self.use_text:
            self.text = TextModule(cfg.text_emb_dim, cfg.text_proj_dim, p=cfg.dropout)
            modal_dims = [cfg.hidden_dim, cfg.hidden_dim, cfg.text_proj_dim]
        else:
            self.text = None
            modal_dims = [cfg.hidden_dim, cfg.hidden_dim]

        self.final_fusion = TraitWiseFusion(modal_dims, cfg.num_traits)
        self.visual_head = nn.Linear(cfg.hidden_dim, cfg.num_traits)
        self.audio_head = nn.Linear(cfg.hidden_dim, cfg.num_traits)
        self.text_head = nn.Linear(cfg.text_proj_dim, cfg.num_traits) if self.use_text else None
        self.bin_head = BinHead(sum(modal_dims), num_traits=cfg.num_traits, num_bins=cfg.num_bins)

    def maybe_modality_dropout(self, modal_vecs):
        if not self.training or cfg.modality_dropout_p <= 0:
            return modal_vecs
        outs = []
        for vec in modal_vecs:
            mask = (torch.rand(vec.shape[0], 1, device=vec.device) >= cfg.modality_dropout_p).float()
            outs.append(vec * mask)
        return outs

    def forward(self, batch):
        v_clip, v_tokens, vis_info = self.visual(batch["faces"], batch["face_scores"], batch["window_mask"])
        a_clip, a_tokens, aud_info = self.audio(batch["mel_windows"], batch["prosody"], batch["ege_maps"], batch["window_mask"])
        av_clip, av_tokens, av_attn = self.av_fusion(v_tokens, a_tokens, batch["window_mask"])

        pred_v = torch.sigmoid(self.visual_head(v_clip))
        pred_a = torch.sigmoid(self.audio_head(a_clip))

        modal_vecs = [av_clip, (v_clip + a_clip) / 2.0]
        pred_t = None
        if self.use_text:
            t_clip = self.text(batch["text_emb"])
            pred_t = torch.sigmoid(self.text_head(t_clip))
            modal_vecs.append(t_clip)
        else:
            t_clip = None

        modal_vecs = self.maybe_modality_dropout(modal_vecs)
        pred, modal_preds, alpha = self.final_fusion(modal_vecs)
        bin_logits = self.bin_head(torch.cat(modal_vecs, dim=-1))

        return {
            "pred": pred,
            "pred_visual": pred_v,
            "pred_audio": pred_a,
            "pred_text": pred_t,
            "bin_logits": bin_logits,
            "v_clip": v_clip,
            "a_clip": a_clip,
            "av_clip": av_clip,
            "text_clip": t_clip,
            "fusion_modal_vecs": modal_vecs,
            "window_tokens_visual": v_tokens,
            "window_tokens_audio": a_tokens,
            "window_tokens_fused": av_tokens,
            "vis_info": vis_info,
            "aud_info": aud_info,
            "fusion_info": {"alpha": alpha, "av_window_attn": av_attn},
        }

sample_prosody_dim = train_ds[0]["prosody"].shape[0]
sample_ege_dim = train_ds[0]["ege_maps"].shape[0]
model = AVTPersonalityModel(vssd_backbone, visual_dim, sample_prosody_dim, sample_ege_dim).to(DEVICE)

batch0 = next(iter(train_loader))
with torch.no_grad():
    batch0_cuda = {k: (v.to(DEVICE) if isinstance(v, torch.Tensor) else v) for k, v in batch0.items()}
    out0 = model(batch0_cuda)
print("pred shape:", out0["pred"].shape)
```


## Losses, metrics и cross-batch memory


```
BIN_BOUNDS = [0.0, 0.4, 0.6, 1.000001]
huber_loss_fn = nn.SmoothL1Loss(beta=cfg.huber_beta)
ce_loss_fn = nn.CrossEntropyLoss(label_smoothing=0.03)

def move_batch_to_device(batch, device):
    out = {}
    for k, v in batch.items():
        out[k] = v.to(device, non_blocking=True) if isinstance(v, torch.Tensor) else v
    return out

def target_to_bins(y):
    borders = torch.tensor(BIN_BOUNDS[1:-1], device=y.device)
    return torch.bucketize(y, boundaries=borders).long()

def tensor_ccc(pred, target, eps=1e-8):
    mx = pred.mean(dim=0)
    my = target.mean(dim=0)
    vx = pred.var(dim=0, unbiased=False)
    vy = target.var(dim=0, unbiased=False)
    cov = ((pred - mx) * (target - my)).mean(dim=0)
    ccc = (2 * cov) / (vx + vy + (mx - my) ** 2 + eps)
    return ccc

def mean_ccc(pred, target):
    return tensor_ccc(pred, target).mean()

def rank_loss(pred, target, ignore_delta=0.02):
    losses = []
    for t in range(target.shape[1]):
        y = target[:, t]
        p = pred[:, t]
        diff_y = y.unsqueeze(1) - y.unsqueeze(0)
        diff_p = p.unsqueeze(1) - p.unsqueeze(0)
        sign = torch.sign(diff_y)
        valid = (torch.abs(diff_y) > ignore_delta).float()
        pair_loss = F.softplus(-sign * diff_p) * valid
        denom = valid.sum().clamp_min(1.0)
        losses.append(pair_loss.sum() / denom)
    return torch.stack(losses).mean()

class ContrastMemory:
    def __init__(self, max_size=2048):
        self.max_size = max_size
        self.emb = None
        self.bins = None

    def clear(self):
        self.emb = None
        self.bins = None

    @torch.no_grad()
    def update(self, emb, bins):
        emb = F.normalize(emb.detach(), dim=-1)
        bins = bins.detach()
        if self.emb is None:
            self.emb = emb
            self.bins = bins
        else:
            self.emb = torch.cat([self.emb, emb], dim=0)[-self.max_size:]
            self.bins = torch.cat([self.bins, bins], dim=0)[-self.max_size:]

def contrastive_loss_with_memory(emb, y, memory: Optional[ContrastMemory], temperature=0.10):
    emb = F.normalize(emb, dim=-1)
    bins = target_to_bins(y)

    if memory is None or memory.emb is None:
        bank_emb = emb
        bank_bins = bins
    else:
        bank_emb = torch.cat([emb, memory.emb.to(emb.device)], dim=0)
        bank_bins = torch.cat([bins, memory.bins.to(emb.device)], dim=0)

    logits = emb @ bank_emb.T / temperature
    losses = []

    for t in range(y.shape[1]):
        pos_mask = (bins[:, t].unsqueeze(1) == bank_bins[:, t].unsqueeze(0)).float()
        if bank_emb.shape[0] >= emb.shape[0]:
            pos_mask[:, :emb.shape[0]] -= torch.eye(emb.shape[0], device=emb.device)
        pos_mask = pos_mask.clamp_min(0.0)

        log_prob = logits - torch.logsumexp(logits, dim=1, keepdim=True)
        denom = pos_mask.sum(dim=1).clamp_min(1.0)
        loss_t = -(pos_mask * log_prob).sum(dim=1) / denom
        losses.append(loss_t.mean())

    return torch.stack(losses).mean()

def compute_loss(outputs, target, contrast_memory=None):
    pred = outputs["pred"]
    pred_v = outputs["pred_visual"]
    pred_a = outputs["pred_audio"]

    loss_main = huber_loss_fn(pred, target)
    loss_aux_v = huber_loss_fn(pred_v, target)
    loss_aux_a = huber_loss_fn(pred_a, target)

    loss_aux_t = pred.new_tensor(0.0)
    if outputs.get("pred_text") is not None:
        loss_aux_t = huber_loss_fn(outputs["pred_text"], target)

    loss_ccc = 1.0 - mean_ccc(pred, target)

    bins = target_to_bins(target)
    logits = outputs["bin_logits"]
    loss_bin = 0.0
    for t in range(cfg.num_traits):
        loss_bin = loss_bin + ce_loss_fn(logits[:, t, :], bins[:, t])
    loss_bin = loss_bin / cfg.num_traits

    emb = torch.cat([
        outputs["av_clip"],
        outputs["text_clip"] if outputs.get("text_clip") is not None else outputs["a_clip"]
    ], dim=-1)
    loss_contrast = contrastive_loss_with_memory(emb, target, contrast_memory, temperature=cfg.contrast_temperature)
    loss_rank = rank_loss(pred, target, ignore_delta=cfg.rank_ignore_delta)

    total = (
        loss_main
        + cfg.lambda_aux_modal * (loss_aux_v + loss_aux_a + loss_aux_t)
        + cfg.lambda_ccc * loss_ccc
        + cfg.lambda_bin * loss_bin
        + cfg.lambda_contrast * loss_contrast
        + cfg.lambda_rank * loss_rank
    )

    return total, {
        "loss_total": total.detach(),
        "loss_main": loss_main.detach(),
        "loss_aux_v": loss_aux_v.detach(),
        "loss_aux_a": loss_aux_a.detach(),
        "loss_aux_t": loss_aux_t.detach(),
        "loss_ccc": loss_ccc.detach(),
        "loss_bin": loss_bin.detach(),
        "loss_contrast": loss_contrast.detach(),
        "loss_rank": loss_rank.detach(),
    }

@torch.no_grad()
def compute_metrics_from_numpy(preds, targets):
    mae_per_trait = np.mean(np.abs(preds - targets), axis=0)
    macc_per_trait = 1.0 - mae_per_trait
    cccs = []
    for i in range(targets.shape[1]):
        x = preds[:, i]
        y = targets[:, i]
        mx, my = x.mean(), y.mean()
        vx, vy = x.var(), y.var()
        cov = ((x - mx) * (y - my)).mean()
        ccc = (2 * cov) / (vx + vy + (mx - my) ** 2 + 1e-8)
        cccs.append(ccc)

    out = {
        "mae_mean": float(mae_per_trait.mean()),
        "macc_mean": float(macc_per_trait.mean()),
        "ccc_mean": float(np.mean(cccs)),
    }
    for i, trait in enumerate(cfg.trait_names):
        out[f"mae_{trait}"] = float(mae_per_trait[i])
        out[f"macc_{trait}"] = float(macc_per_trait[i])
        out[f"ccc_{trait}"] = float(cccs[i])
    return out
```


## EMA, training loop и staged unfreezing


```
class ModelEMA:
    def __init__(self, model, decay=0.999):
        self.decay = decay
        self.ema = copy.deepcopy(model).eval()
        for p in self.ema.parameters():
            p.requires_grad_(False)

    @torch.no_grad()
    def update(self, model):
        msd = model.state_dict()
        for k, v in self.ema.state_dict().items():
            if v.dtype.is_floating_point:
                v.copy_(v * self.decay + msd[k].detach() * (1.0 - self.decay))
            else:
                v.copy_(msd[k])

def freeze_module(module):
    for p in module.parameters():
        p.requires_grad = False

def unfreeze_module(module):
    for p in module.parameters():
        p.requires_grad = True

def set_backbone_train_mode(model, mode="frozen"):
    freeze_module(model.visual.backbone)

    if mode == "frozen":
        pass
    elif mode == "last_stage":
        if hasattr(model.visual.backbone, "layers"):
            unfreeze_module(model.visual.backbone.layers[-1])
            if hasattr(model.visual.backbone, "norm"):
                unfreeze_module(model.visual.backbone.norm)
    elif mode == "all":
        unfreeze_module(model.visual.backbone)
    else:
        raise ValueError(mode)

    trainable_modules = [
        model.visual.frame_norm, model.visual.frame_pool, model.visual.window_proj, model.visual.bigru,
        model.visual.self_attn, model.visual.temporal_ln, model.visual.window_pool, model.visual.out,
        model.audio, model.av_fusion, model.final_fusion, model.visual_head, model.audio_head, model.bin_head
    ]
    if model.use_text:
        trainable_modules.extend([model.text, model.text_head])

    for m in trainable_modules:
        unfreeze_module(m)

def make_optimizer(model, lr_heads, lr_backbone):
    backbone_params = []
    head_params = []
    for name, p in model.named_parameters():
        if not p.requires_grad:
            continue
        if name.startswith("visual.backbone"):
            backbone_params.append(p)
        else:
            head_params.append(p)

    param_groups = []
    if head_params:
        param_groups.append({"params": head_params, "lr": lr_heads, "weight_decay": cfg.weight_decay})
    if backbone_params:
        param_groups.append({"params": backbone_params, "lr": lr_backbone, "weight_decay": cfg.weight_decay})

    return torch.optim.AdamW(param_groups, betas=(0.9, 0.999))

def make_scheduler(optimizer, total_steps, warmup_ratio=0.08):
    warmup_steps = max(1, int(total_steps * warmup_ratio))
    def lr_lambda(step):
        if step < warmup_steps:
            return step / warmup_steps
        progress = (step - warmup_steps) / max(1, total_steps - warmup_steps)
        return 0.5 * (1.0 + math.cos(math.pi * progress))
    return torch.optim.lr_scheduler.LambdaLR(optimizer, lr_lambda)

def run_epoch(model, loader, optimizer=None, scheduler=None, scaler=None, train=True, ema=None):
    model.train() if train else model.eval()

    all_preds = []
    all_targets = []
    loss_rows = []
    contrast_memory = ContrastMemory(max_size=cfg.contrast_memory_size) if train else None

    if optimizer is not None:
        optimizer.zero_grad(set_to_none=True)

    for step, batch in enumerate(tqdm(loader, total=len(loader))):
        if batch is None:
            continue

        batch = move_batch_to_device(batch, DEVICE)
        with torch.set_grad_enabled(train):
            with torch.cuda.amp.autocast(enabled=cfg.amp):
                outputs = model(batch)
                loss, loss_dict = compute_loss(outputs, batch["target"], contrast_memory=contrast_memory)

            if train:
                loss_scaled = loss / cfg.grad_accum_steps
                if scaler is not None:
                    scaler.scale(loss_scaled).backward()
                else:
                    loss_scaled.backward()

                bins = target_to_bins(batch["target"])
                contrast_memory.update(
                    torch.cat([
                        outputs["av_clip"],
                        outputs["text_clip"] if outputs.get("text_clip") is not None else outputs["a_clip"]
                    ], dim=-1),
                    bins
                )

                if ((step + 1) % cfg.grad_accum_steps == 0) or (step + 1 == len(loader)):
                    if scaler is not None:
                        scaler.unscale_(optimizer)
                    torch.nn.utils.clip_grad_norm_(model.parameters(), cfg.max_grad_norm)

                    if scaler is not None:
                        scaler.step(optimizer)
                        scaler.update()
                    else:
                        optimizer.step()

                    optimizer.zero_grad(set_to_none=True)
                    if scheduler is not None:
                        scheduler.step()
                    if ema is not None:
                        ema.update(model)

        all_preds.append(outputs["pred"].detach().cpu().numpy())
        all_targets.append(batch["target"].detach().cpu().numpy())
        loss_rows.append({k: float(v.detach().cpu()) for k, v in loss_dict.items()})

    preds = np.concatenate(all_preds, axis=0)
    targets = np.concatenate(all_targets, axis=0)

    losses = pd.DataFrame(loss_rows).mean().to_dict()
    metrics = compute_metrics_from_numpy(preds, targets)
    return {**losses, **metrics}, preds, targets

def checkpoint_payload(model, optimizer, scheduler, epoch, stage_name, metrics, ema=None):
    return {
        "epoch": epoch,
        "stage_name": stage_name,
        "model_state_dict": model.state_dict(),
        "optimizer_state_dict": optimizer.state_dict() if optimizer is not None else None,
        "scheduler_state_dict": scheduler.state_dict() if scheduler is not None else None,
        "metrics": metrics,
        "cfg": asdict(cfg),
        "ema_state_dict": ema.ema.state_dict() if ema is not None else None,
    }

def save_checkpoint(payload, path):
    torch.save(payload, path)

def fit_stage(model, train_loader, val_loader, stage_name, epochs, lr_heads, lr_backbone, backbone_mode):
    set_backbone_train_mode(model, backbone_mode)
    optimizer = make_optimizer(model, lr_heads=lr_heads, lr_backbone=lr_backbone)
    steps_per_epoch = math.ceil(len(train_loader) / max(1, cfg.grad_accum_steps))
    total_steps = max(1, epochs * steps_per_epoch)
    scheduler = make_scheduler(optimizer, total_steps=total_steps)
    scaler = torch.cuda.amp.GradScaler(enabled=cfg.amp)
    ema = ModelEMA(model, decay=cfg.ema_decay) if cfg.use_ema else None

    history = []
    best_items = []

    for epoch in range(1, epochs + 1):
        print(f"\n{stage_name}: epoch {epoch}/{epochs}")
        tr_metrics, _, _ = run_epoch(model, train_loader, optimizer=optimizer, scheduler=scheduler, scaler=scaler, train=True, ema=ema)

        eval_model = ema.ema if ema is not None else model
        va_metrics, _, _ = run_epoch(eval_model, val_loader, train=False)

        row = {"stage": stage_name, "epoch": epoch}
        row.update({f"tr_{k}": v for k, v in tr_metrics.items()})
        row.update({f"va_{k}": v for k, v in va_metrics.items()})
        history.append(row)

        score = 0.75 * va_metrics["ccc_mean"] + 0.25 * va_metrics["macc_mean"]
        print({
            "val_ccc_mean": va_metrics["ccc_mean"],
            "val_macc_mean": va_metrics["macc_mean"],
            "val_mae_mean": va_metrics["mae_mean"],
        })

        path = os.path.join(cfg.checkpoints_dir, f"{stage_name}_epoch{epoch}.pt")
        payload = checkpoint_payload(model, optimizer, scheduler, epoch, stage_name, va_metrics, ema=ema)
        save_checkpoint(payload, path)

        best_items.append((score, path))
        best_items = sorted(best_items, key=lambda x: x[0], reverse=True)[:cfg.keep_top_k]

    hist = pd.DataFrame(history)
    hist.to_csv(os.path.join(cfg.logs_dir, f"{stage_name}_history.csv"), index=False)

    top_paths = [p for _, p in best_items]
    print("top checkpoints:", top_paths)
    return hist, top_paths

def load_checkpoint_into_model(model, ckpt_path, use_ema=True, strict=True):
    obj = torch.load(ckpt_path, map_location="cpu", weights_only=False)
    sd = obj["ema_state_dict"] if use_ema and obj.get("ema_state_dict") is not None else obj["model_state_dict"]
    model.load_state_dict(sd, strict=strict)
    return obj
```

```
import os
import torch
import pandas as pd
from tqdm.auto import tqdm

def is_valid_cache(path: str):
    if not os.path.exists(path):
        return False, "missing"
    try:
        _ = torch.load(path, map_location="cpu", weights_only=False)
        return True, None
    except Exception as e:
        return False, repr(e)

def filter_manifest_to_valid_cache(manifest: pd.DataFrame, mode: str, verbose_name: str):
    good_rows = []
    bad_rows = []

    for _, row in tqdm(manifest.iterrows(), total=len(manifest), desc=f"check {verbose_name}"):
        cache_path = get_cache_path(row["split"], row["video_id"], mode)
        ok, err = is_valid_cache(cache_path)
        if ok:
            good_rows.append(row)
        else:
            bad_rows.append({
                "video_id": row["video_id"],
                "cache_path": cache_path,
                "error": err
            })

    good_df = pd.DataFrame(good_rows).reset_index(drop=True)
    bad_df = pd.DataFrame(bad_rows)

    print(f"{verbose_name}: valid {len(good_df)} / {len(manifest)}")
    print(f"{verbose_name}: bad   {len(bad_df)}")

    if len(bad_df) > 0:
        display(bad_df.head(20))

    return good_df, bad_df
```

```
@torch.no_grad()
def predict_loader(model, loader, flip_tta=False):
    model.eval()
    video_ids = []
    preds = []
    targets = []

    for batch in tqdm(loader, total=len(loader)):
        video_ids.extend(batch["video_id"])
        targets.append(batch["target"].numpy())
        batch = move_batch_to_device(batch, DEVICE)

        with torch.cuda.amp.autocast(enabled=cfg.amp):
            out = model(batch)
            pred = out["pred"]

            if flip_tta:
                batch_flip = dict(batch)
                batch_flip["faces"] = torch.flip(batch["faces"], dims=[-1])
                out_flip = model(batch_flip)
                pred = 0.5 * (pred + out_flip["pred"])

        preds.append(pred.detach().cpu().numpy())

    preds = np.concatenate(preds, axis=0)
    targets = np.concatenate(targets, axis=0)
    return video_ids, preds, targets

def save_predictions_csv(video_ids, preds, out_path):
    df = pd.DataFrame(preds, columns=list(cfg.trait_names))
    df.insert(0, "video_id", video_ids)
    df.to_csv(out_path, index=False)
    return df

def ensemble_predict(checkpoint_paths, loader, flip_tta=False):
    preds_list = []
    targets_ref = None
    ids_ref = None

    for ckpt in checkpoint_paths:
        tmp_model = AVTPersonalityModel(vssd_backbone, visual_dim, sample_prosody_dim, sample_ege_dim).to(DEVICE)
        _ = load_checkpoint_into_model(tmp_model, ckpt, use_ema=True, strict=True)
        ids, preds, targets = predict_loader(tmp_model, loader, flip_tta=flip_tta)
        preds_list.append(preds)
        if ids_ref is None:
            ids_ref = ids
            targets_ref = targets
        del tmp_model
        gc.collect()
        if torch.cuda.is_available():
            torch.cuda.empty_cache()

    preds_ens = np.mean(preds_list, axis=0)
    return ids_ref, preds_ens, targets_ref

def evaluate_split(checkpoint_paths, manifest, mode="eval", split_name="validation", flip_tta=True):
    _, loader = make_loader(manifest, mode=mode, shuffle=False)
    ids, preds, targets = ensemble_predict(checkpoint_paths, loader, flip_tta=flip_tta)
    metrics = compute_metrics_from_numpy(preds, targets)
    out_csv = os.path.join(cfg.logs_dir, f"{split_name}_{mode}_predictions.csv")
    df = save_predictions_csv(ids, preds, out_csv)
    return metrics, df

# обычная валидация
val_metrics, val_pred_df = evaluate_split(top_checkpoints, val_manifest, mode="eval", split_name="validation", flip_tta=cfg.use_flip_tta)
print("VAL:", val_metrics)
display(val_pred_df.head())

# dense validation — теперь это реально другой cache signature
dense_val_metrics, dense_val_pred_df = evaluate_split(top_checkpoints, val_manifest, mode="dense", split_name="validation", flip_tta=cfg.use_flip_tta)
print("DENSE VAL:", dense_val_metrics)

# test
# если annotation_test.pkl содержит ground-truth, метрики посчитаются.
# если это скрытый test без истинных label, просто будет сохранён CSV с предсказаниями.
test_metrics, test_pred_df = evaluate_split(top_checkpoints, test_manifest, mode="test", split_name="test", flip_tta=cfg.use_flip_tta)
print("TEST:", test_metrics)
display(test_pred_df.head())
```

```
def make_saliency_map(model, batch, sample_idx=0, trait_idx=0):
    model.eval()
    faces = batch["faces"].clone().detach().to(DEVICE)
    faces.requires_grad_(True)

    new_batch = {
        "video_id": batch["video_id"],
        "faces": faces,
        "face_scores": batch["face_scores"].to(DEVICE),
        "mel_windows": batch["mel_windows"].to(DEVICE),
        "window_mask": batch["window_mask"].to(DEVICE),
        "prosody": batch["prosody"].to(DEVICE),
        "ege_maps": batch["ege_maps"].to(DEVICE),
        "text_emb": batch["text_emb"].to(DEVICE),
        "target": batch["target"].to(DEVICE),
    }

    model.zero_grad(set_to_none=True)
    with torch.backends.cudnn.flags(enabled=False):
        out = model(new_batch)
        score = out["pred"][sample_idx, trait_idx]
        score.backward()

    grads = faces.grad[sample_idx].abs().mean(dim=1)
    sal = grads.detach().cpu().numpy()
    return sal

batch0 = next(iter(val_loader))
sal = make_saliency_map(model, batch0, sample_idx=0, trait_idx=2)
plt.figure(figsize=(10, 4))
for i in range(min(6, sal.shape[0])):
    plt.subplot(2, 3, i + 1)
    plt.imshow(sal[i, 0], cmap="hot")
    plt.axis("off")
plt.tight_layout()
plt.show()
```
